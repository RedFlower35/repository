```python
def print_christmas_tree():
    """
    印出一棵聖誕樹。
    這個函式沒有參數，也沒有回傳值，會直接用 print() 畫出聖誕樹的圖案。
    """
    height = 5
    for i in range(height):
        print(' ' * (height - i - 1) + '*' * (2 * i + 1))
    print(' ' * (height - 1) + '|')
```

```python
def repeat_string(s, n):
    """
    傳回字串 s 重複 n 次的結果。
    參數:
        s (str): 要重複的字串
        n (int): 重複的次數
    回傳:
        str: 重複後的字串
    """
    return s * n
```

```python
# 自訂的 function，有參數，有傳出值
# todo: 建立一個簡單問候函數
# 名稱: greet_person(snake_case)
# 參數: name, greeting, greeting 有預設值
```

```python
def greet_person(name, greeting="Hello"):
    """
    傳回簡單的問候語。
    參數:
        name (str): 被問候的人名
        greeting (str, 選填): 問候語，預設為 "Hello"
    回傳:
        str: 組合後的問候語句
    """
    return f"{greeting}, {name}!"
```

```python
# todo: 建立一個BMI計算機
# 名稱: calculate_bmi
# 參數: weight, height
# 傳出值: (tuple):(bmi, status)
```

```python
def calculate_bmi(weight, height):
    """
    計算 BMI 並回傳 BMI 值與健康狀態。
    參數:
        weight (float): 體重 (公斤)
        height (float): 身高 (公尺)
    回傳:
        tuple: (bmi, status)
            bmi (float): 計算出的 BMI 值
            status (str): 健康狀態描述
    """
    bmi = weight / (height ** 2)
    if bmi < 18.5:
        status = "過輕"
    elif bmi < 24:
        status = "正常"
    elif bmi < 27:
        status = "過重"
    else:
        status = "肥胖"
    return bmi, status
```

```python
def calculate_bmi(weight, height_cm):
    """
    計算 BMI 並回傳 BMI 值與健康狀態。
    參數:
        weight (float): 體重 (公斤)
        height_cm (float): 身高 (公分)
    回傳:
        tuple: (bmi, status)
            bmi (float): 計算出的 BMI 值
            status (str): 健康狀態描述
    """
    height_m = height_cm / 100
    bmi = weight / (height_m ** 2)
    if bmi < 18.5:
        status = "過輕"
    elif bmi < 24:
        status = "正常"
    elif bmi < 27:
        status = "過重"
    else:
        status = "肥胖"
    return bmi, status
```

```python
# 範例：呼叫上面定義的函式

# 呼叫 print_christmas_tree() 印出聖誕樹
print_christmas_tree()

# 呼叫 repeat_string() 重複字串
result = repeat_string("Hi! ", 3)
print(result)

# 呼叫 greet_person() 問候
greeting = greet_person("小明")
print(greeting)

# 呼叫 calculate_bmi() 計算 BMI
bmi, status = calculate_bmi(70, 175)
print(f"BMI: {bmi:.2f}, 狀態: {status}")
```

```python
def password_strength(password):
    """
    驗證密碼強度。
    參數:
        password (str): 要驗證的密碼
    回傳:
        str: 密碼強度描述 ("強", "中等", "弱")
    """
    if len(password) < 8:
        return "密碼強度: 弱"
    has_digit = any(c.isdigit() for c in password)
    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    if has_digit and has_upper and has_lower:
        return "密碼強度: 強"
    elif (has_upper and has_lower) or (has_digit and (has_upper or has_lower)):
        return "密碼強度: 中等"
    else:
        return "密碼強度: 弱"
```
